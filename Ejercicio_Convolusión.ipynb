```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de los arreglos dados
h_u = [7, 0.5, 8, -2, 0.3]  # El valor 8 corresponde a n=0
x_n = [10, -8, 3.5, 6]       # El valor -8 corresponde a n=0
print("h_u:", h_u)
print("x_n:", x_n)

# Representación en un sistema indexado con origen en n=0
h_n = np.array(h_u)
x_n = np.array(x_n)
print("h_n (numpy array):", h_n)
print("x_n (numpy array):", x_n)

# Realizando la convolución entre h[n] y x[n]
y_n = np.convolve(h_n, x_n)
print("y_n (resultado de la convolución):", y_n)

# Graficando la salida Y[n]
n_y = np.arange(-(len(h_n) - 1), len(x_n))
print("n_y (eje de índices para Y[n]):", n_y)

plt.stem(n_y, y_n, basefmt=" ")
plt.xlabel('n')
plt.ylabel('Y[n]')
plt.title('Respuesta del sistema Y[n] = H{X[n]}')
plt.grid()
plt.show()
```
